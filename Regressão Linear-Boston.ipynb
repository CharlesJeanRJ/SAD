### Prepare X e y usando o pandas

```python
feature_cols = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']

X = dfboston[feature_cols]

X = dfboston[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']]

X.head()
```

```python
y = dfboston['target']

y = dfboston.target

y.head()
```

### Gere um X de treino e y de treino, X de teste e y da base (Split Train/Test)

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

### Qual o percentual gerado para criar o conjunto de treino e o conjunto de teste?

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

### Conjunto de Treino = 74,9%
### Conjunto de Teste = 25,1%

### Usando modelo de regressão linear do sklearn, treine o modelo com o X e y de treino

```python
from sklearn.linear_model import LinearRegression

linreg = LinearRegression()

linreg.fit(X_train, y_train)
```

### Imprima os atributos de "intercept" e "coefficients" que foram gerados

```python
print(linreg.intercept_)
```

```python
print(linreg.coef_)
```

### Imprima o par "feature names" com os "coefficients"

```python
list(zip(feature_cols, linreg.coef_))
```

### Faça uma previsão usando o conjunto de X de teste

```python
y_pred = linreg.predict(X_test)
```

### Mostre a margem de error, usando o método "Root Mean Squared Error (RMSE)"

```python
from sklearn import metrics
import numpy as np
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

### Existe uma forma de melhorar essa margem para que fique menor? Se sim, como seria?

### Sim, estudando os dados dos gráficos de features e responses e retirando as features onde existe bastante dados distantes das retas. Ex: retirando o CRIM.

```python
feature_cols = ['ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']

X = dfboston[feature_cols]

y = dfboston.target

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

linreg.fit(X_train, y_train)

y_pred = linreg.predict(X_test)

print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
